## Matplotlib example workflow

```python
# 0. Import matplotlib and get it ready for plotting in Jupyter
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare data
x = [1,2,3,4]
y = [11,22,27,44]

# 2. Setup plot
fig, ax = plt.subplots(figsize = (10, 5)) # (width, height)

# 3. Plot data
ax.plot(x, y)

# 4. Customize plot
ax.set(title = "Simple Plot", xlabel = "x-axis", ylabel = "y-axis")

# 5. Save & show
fig.savefig("images/sample-plot.png")
```

```python
import numpy as np
```

```python
# Create some data
x = np.linspace(0, 10, 100)
x[:10]
```

```python
# Plot the data
fig, ax = plt.subplots()
ax.plot(x, x**2);
```

```python
# Make a plot from dictionary
nut_butter_prices = {"A": 10, "B": 8, "C": 12}
fig, ax = plt.subplots()
ax.set(title="Nut butter prices", ylabel="Price", xlabel="Name")
ax.bar(nut_butter_prices.keys(), nut_butter_prices.values());
```

```python
fig, ax = plt.subplots()
ax.set(title="Nut butter prices")
ax.barh(list(nut_butter_prices.keys()), list(nut_butter_prices.values()));
```

```python
# Make some data for histograms and plot it
x = np.random.randn(1000)
fig, ax = plt.subplots()
ax.hist(x);
```

### Two options for subplots

```python
# Subplot option 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows = 2, ncols = 2, figsize = (10, 10))
ax1.set(title = "Plot N1", xlabel="Price", ylabel="Name")
ax2.set(title = "Plot N2")
ax3.set(title = "Plot N3")
ax4.set(title = "Plot N4")
ax1.barh(list(nut_butter_prices.keys()), list(nut_butter_prices.values()));
ax2.plot(y);
ax3.plot(np.random.randint(1,400,30));
ax4.scatter(np.random.random(10), np.random.random(10));
```

```python
# Subplot option 2
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(10,10))
# Plto to each different index
ax[0,0].set(title = "Plot N1", xlabel="Price", ylabel="Name")
ax[0,1].set(title = "Plot N2")
ax[1,0].set(title = "Plot N3")
ax[1,1].set(title = "Plot N4")
ax[0,0].barh(list(nut_butter_prices.keys()), list(nut_butter_prices.values()));
ax[0,1].plot(y);
ax[1,0].plot(np.random.randint(1,400,30));
ax[1,1].scatter(np.random.random(100), np.random.random(100));
```

```python
# Trying an another dataset
heart_disease = pd.read_csv("heart-disease.csv")
heart_disease.head()
```

```python
heart_disease["age"].plot.hist(bins = 20);
```

```python
heart_disease.head()
```

```python
heart_disease.plot.hist(figsize=(10,60),subplots=True);
```

```python
over_50 = heart_disease[heart_disease["age"] > 50]
over_50.head()
```

```python
#Pyplot method
over_50.plot(kind="scatter", x='age', y='chol', c='target');
```

```python
## OO method mixed with pyplot method
fig, ax = plt.subplots(figsize=(10,6))
over_50.plot(kind='scatter', x='age', y='chol', c='target', ax=ax);
ax.set_xlim([45,80]);
```

```python
# OO method from scratch
fig, ax = plt.subplots(figsize=(11,6))

# Plot the data
scatter = ax.scatter(x=over_50["age"], 
                     y=over_50["chol"], 
                     c=over_50["target"]);

# Customize
ax.set(title = "Heart Disease and Cholesterol Levels", 
       xlabel="Age", ylabel="Cholesterol");

# Add a legend
ax.legend(*scatter.legend_elements(), title="Target");

# Add a horizontal line
ax.axhline(over_50["chol"].mean(), linestyle=":", color="red");
ax.axvline(over_50["age"].mean(), linestyle="--", color="blue")
```

```python
over_50
```

```python
# Subplot of chol, age, thalach
fig, (ax0, ax1) = plt.subplots(figsize=(10,14), nrows=2, ncols=1, sharex=True)

# Add data to ax0
scatter = ax0.scatter(x=over_50["age"], 
                      y=over_50["chol"],
                      c=over_50["target"])
# Customize ax0
ax0.set(title="Age and Cholesterol Level",
       ylabel="Cholesterol")
ax0.axhline(y=over_50["chol"].mean(), linestyle="--")
ax0.axvline(x=over_50["age"].mean(), linestyle=":")
ax0.legend(*scatter.legend_elements(), title="Target");


# Add data to ax1
scatter = ax1.scatter(y=over_50["thalach"],
                     x=over_50["age"],
                     c=over_50["target"])
# Customize ax1
ax1.set(title="Max Heart Rate and Age",
       ylabel="Max Heart Rate", xlabel="Age")
ax1.axvline(x=over_50["age"].mean(), linestyle="--", color="red")
ax1.axhline(y=over_50["thalach"].mean(), linestyle=":", color="red")
ax1.legend(*scatter.legend_elements(), title="Target");

#Add a tite to the figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold")
```

```python
plt.style.use("seaborn-v0_8-whitegrid")

# Subplot of chol, age, thalach
fig, (ax0, ax1) = plt.subplots(figsize=(10,14), nrows=2, ncols=1, sharex=True)

# Add data to ax0
scatter = ax0.scatter(x=over_50["age"], 
                      y=over_50["chol"],
                      c=over_50["target"],
                     cmap="winter")
# Customize ax0
ax0.set(title="Age and Cholesterol Level",
       ylabel="Cholesterol")
ax0.axhline(y=over_50["chol"].mean(), linestyle="--", color="blue")
ax0.axvline(x=over_50["age"].mean(), linestyle=":", color="blue")
ax0.legend(*scatter.legend_elements(), title="Target");


# Add data to ax1
scatter = ax1.scatter(y=over_50["thalach"],
                     x=over_50["age"],
                     c=over_50["target"],
                     cmap="plasma")
# Customize ax1
ax1.set(title="Max Heart Rate and Age",
       ylabel="Max Heart Rate", xlabel="Age")
ax1.axvline(x=over_50["age"].mean(), linestyle="--", color="red")
ax1.axvline(x=over_50["age"].max(), linestyle="--", color="green")
ax1.axvline(x=over_50["age"].min(), linestyle="--", color="blue")
ax1.axhline(y=over_50["thalach"].mean(), linestyle=":", color="red")
ax1.axhline(y=over_50["thalach"].max(), linestyle=":", color="green")
ax1.axhline(y=over_50["thalach"].min(), linestyle=":", color="blue")
ax1.legend(*scatter.legend_elements(), title="Target");

#Add a tite to the figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");
```
